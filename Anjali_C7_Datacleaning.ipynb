```python
def analyze_continue_categorical_columns(df):
    
    df_conti = df.select_dtypes(include="number")
    df_cate = df.select_dtypes(exclude="number")
    
    print("Continuous columns are:",df_conti.columns.tolist())
    print("Categorical columns are:",df_cate.columns.tolist())
    
analyze_continue_categorical_columns(df)  
```

```python
df_cate = df.select_dtypes(exclude="number")
print("Categorical columns are:", df_cate.columns.tolist())

for column in df_cate.columns:
        df[column] = df[column].fillna("Not Available")
```

```python
df_conti = df.select_dtypes(include="number")
print("Continuous columns are:",df_conti.columns.tolist())

for column in df_conti.columns:
    median_value = df[column].median()  # Calculate the median
    df[column].fillna(median_value, inplace=True) 
```
